It is important to distinguish between migrant stocks and migration flows. These numbers are so-called “absolute indicators of migration”. They show how many migrants reside in a country as of a given date, i.e. at a certain moment, and how many migration events or procedures associated with migration have occurred within a period of time. \

Over the past 30 years, the percentage of people over 65 in Finland has risen from 14% to 23%, and the aged dependency ratio* increased from 21% to 38%.\
Economists use this ratio to analyze how demographic changes impact economies, with working-age individuals contributing most to the economy and taxes. Societies with high dependency ratios may face challenges, while those with lower ratios are considered better off. Migration is one factor affecting the dependency ratio. Rapidly aging populations create a shortage of qualified workers, leading some countries to turn to immigration to maintain a well-supplied labor force. As we can see migrant stocks improves the situation with aging population.\

**Aged dependency ratio - older than 64 number of persons divided by the working-age population then multiplying by 100.*

Migration is a powerful driver and consequence of change, but measuring it accurately with big data remains a challenge. Differentiating between migration and short-term mobility is unclear, making it difficult to identify true migrants from the data.

### Migrant stocks and flows analysis

Let's have a look at migrant stocks in Finland 1990-2022 (https://public.tableau.com/views/StockofmigrantsinFinland1990-2022/Dashboard?:language=en-US&:display_count=n&:origin=viz_share_link).

```python
image_path = 'C:/Users/dnaadmin/Desktop/project/project plan with photos/stocks.png'

# Read the image using OpenCV
img = cv2.imread(image_path)

# OpenCV reads images in BGR format. Convert it to RGB to display with matplotlib.
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide the axis
plt.show()
```

The interactive Tableau dashboard provides a comprehensive visual representation of the dynamic changes in international migrant stocks in Finland from 1990 to 2022.\

During 30 years Finland witnessed a remarkable eight-fold increase in its migrant population. One particular standout is the Uusimaa region, which experienced an unparalleled surge in migrant numbers, soaring from 37% in 1990 to an impressive 55% in 2022. This enduring trend makes Uusimaa the leading region for migrant settlement in the country. Following closely behind, Southwest Finland has consistently held the second position, while Pirkanmaa has secured the third spot. Additionally, Pohjois-Pohjanmaa and Ostrobothnia regions also stand out for notable migrant presence.\

Intriguingly, the gender distribution among migrants has remained relatively stable across all years, showcasing a consistent balance.\

Looking back three decades, the main source of immigrant stocks in Finland was the rest of Europe, contributing 59% of migrants, followed by Asia at 14%. In contrast, the present-day scenario has seen a shift, with Europe now accounting for 50% of the migrant population, and Asia surging to 32%.\

Discussing flows, it has experienced remarkable growth, nearly tripling in size. Notably, the proportion of immigrants with Finnish heritage has remained relatively stable. However, significant changes have been observed among immigrants with foreign origins, witnessing a remarkable five-fold increase.\ 

Over the past three decades, substantial migration has surged towards the Uusimaa region. Back in 1990, the Southwest region ranked second, with Pirkanmaa and Lapland following suit. Initially, immigrants with Finnish background formed the majority of arrivals. However, after 2006, a shift occurred, with European immigrants consistently dominating the inflow. Towards 2021, a noteworthy portion of immigrants originated from Asia. Throughout all these years, the gender distribution among migrants has remained consistently balanced, exhibiting a stable trend. Emigrant flows from Finland have doubled.
